```python
import requests

def fetch_posts(api_url="https://jsonplaceholder.typicode.com/posts"):
    try:
        with requests.Session() as session:
            response = session.get(api_url)
            posts = response.json()
            return posts
    except requests.exceptions.RequestException as e:
        print(f"Ошибка при запросе: {e}")
        return None
posts = fetch_posts()
if posts:
    print("Список:")
    for post in posts:
        print(f"  ID: {post['id']}, Заголовок: {post['title']}") 
```
